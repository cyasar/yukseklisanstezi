```python
import numpy as np
import matplotlib.pyplot as plt
from qutip import *

# Sinyal alanı durumunu tanımla (örneğin, koherent durum)
alpha = 1.0
signal = coherent(10, alpha)

# Yerel osilatörün fazı
theta = 0.0

# Kuadratur operatörü
quadrature_op = (destroy(10) * np.exp(-1j * theta) + destroy(10).dag() * np.exp(1j * theta)).unit()

# Homodin ölçümünü simüle et
num_measurements = 1000
results = []

for _ in range(num_measurements):
    result = quadrature_op.matrix_element(signal.dag(), signal)
    results.append(np.real(result))

# Ölçüm sonuçlarını yazdır
print("Measurement Results:", results)

# Histogramı çiz
plt.hist(results, bins=30, edgecolor='black', alpha=0.7)
plt.xlabel('Measurement Outcome')
plt.ylabel('Frequency')
plt.title('Homodyne Measurement Outcomes')
plt.show()
```
